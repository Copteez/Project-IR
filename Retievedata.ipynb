```python
from bs4 import BeautifulSoup
import requests
import csv 
import pandas as pd
response = requests.get(url)
print(response)
with open("DOC_LIST.csv",'w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["DOC_ID","Food title", "Food description (Normal)", "Effect(normal)", "Food description (Delicious)", "Effect(Delicious)", "Food description (Suspicious)", "Effect(Suspicious)", "Type", "Region", "Recipe", "Quality", "Proficiency"])
    if response.status_code == 200:
        soup = BeautifulSoup(response.content,'html.parser')
        a_href=soup.findAll("a")
        href = []
        for element in a_href:
            text = (str)(element.get('href'))
            if (text.startswith('/wiki/Special:Search'))or (text.startswith('/wiki/Category:Food')) or (text.startswith('/wiki/Category:Food_with_Recipes?action=history')) or(text.startswith('/wiki/Category_talk:Food_with_Recipes?action=edit&redlink=1'))or(text.startswith('/wiki/Food'))or(text.startswith('/wiki/Cooking'))or(text.startswith('/wiki/Local_Sitemap'))or(text.startswith('/wiki/Special:Categories')):
                continue
            if (text.startswith('/wiki')):
                href.append("https://genshin-impact.fandom.com"+text)
        href = list(dict.fromkeys(href))
        # For check linked
        # print(href) 
        i = -1
        index = 0
        for i in range(150):
            t = i+1
            index = index+1
            sub_URL_rq = requests.get(href[i])
            print(sub_URL_rq)
            if (sub_URL_rq.status_code) == 200:
                sub_soup = BeautifulSoup(sub_URL_rq.content,'html.parser')
                Food_Title = sub_soup.find(class_="mw-page-title-main").getText()
                print(Food_Title)
                Pre_Food_description = sub_soup.findAll(class_ = "pi-data-value pi-font")
                Food_description_List = []
                Food_description_List.append(index)
                Food_description_List.append(Food_Title)
                for text in Pre_Food_description:
                    Food_description_List.append(text.get_text())
                print(Food_description_List)
                writer.writerow(Food_description_List)
            
import pandas as pd
data = pd.read_csv("DOC_LIST.csv",encoding='ISO-8859-1')
data

        
```
